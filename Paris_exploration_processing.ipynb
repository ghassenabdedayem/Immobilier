```python
df[not('PARIS' in df['Commune'])] = df['Commune'] = 'PARIS'
df.head()
```

```python
def str_int(x):
    return str(int(float(x)))

strings = ['No_voie', 'Code_postal']

for i in strings:
    df[i] = df[i].apply(str_int)

df.head()
```

```python
col_adress = ['Type_de_voie', 'Voie', 'Code_postal', 'Commune']

df['complete_address'] = ' '

for i in col_adress:
    df['complete_address'] = df['complete_address'] + ' ' + df[i]

df.head()
```

```python
df.to_csv('../cleaned_paris_vs2.csv', sep='|')
```

```python
def address_long(x):
    location = geolocator.geocode(query=x)
    if location == None:
        return 'Error'
    else:
        return str(location.longitude)

def address_lati(x):
    location = geolocator.geocode(query=x)
    if location == None:
        return 'Error'
    else:
        return str(location.latitude)
```
